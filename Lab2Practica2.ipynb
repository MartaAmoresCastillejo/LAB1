### Primer apartado

Para la columna que contiene las secuencias de genes,   generar una nueva columna (con la que trabajaremos posteriormente). Esta nueva columna debe contener los primeros dígitos de los genes siento este número aleatorio contenido en el intervalo 4-10, es decir, cortaremos las secuencias para que su longitud sea aleatoria entre 4 y 10 dígitos. 

```python
import pandas as pd
import random

# cargar los datos en un dataframe
data = pd.read_csv('splice.data', header=None)

# Obtener la tercera columna del dataframe
data[3] = data[2]

# Función para acortar cadenas de manera aleatoria
def acortar_cadena(cadena):
    length = random.randint(4, 10) # Obtener un número aleatorio entre 4 y 10
    return cadena[:length]

data[3] = acortar_cadena(data[3])
data
```

```python
frase = "Esta es una frase de ejemplo"
frase_acortada = acortar_cadena(frase)
print(frase_acortada)
```

```python
import pandas as pd
import random

# cargar los datos en un dataframe
data = pd.read_csv('splice.data', header=None)


# Definir una función para acortar los elementos
def acortar_elemento(elemento):
    longitud = random.randint(4, 10)
    return elemento[:longitud]

# Recorrer la columna 'ciudad' y acortar sus elementos
for index, row in data.iterrows():
    ciudad = row[data[2]]
    ciudad_acortada = ciudad.apply(acortar_elemento)
    data.at[index, data[2]] = ciudad_acortada

# Imprimir el DataFrame resultante
data
```
